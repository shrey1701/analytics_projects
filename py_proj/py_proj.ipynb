# Exploratory data analysis

```python
vw = sns.countplot(x='Gender', data=dfa)

for bars in vw.containers:
    vw.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

```python
ax=sns.countplot(x='Age Group', data=dfa, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_age = dfa.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
orders_state = dfa.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='State', y='Orders', data=orders_state)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sales_state = dfa.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='State', y='Amount', data=sales_state)
```

```python
mr = sns.countplot(x='Marital_Status', data=dfa)

for bars in mr.containers:
    mr.bar_label(bars)
```

```python
mar_sales = dfa.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Marital_Status', y='Amount', data= mar_sales, hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
pz = sns.countplot(x='Occupation', data=dfa)
for bars in pz.containers:
    pz.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
occ_sales=dfa.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Occupation', y='Amount', data=occ_sales)
```

```python
sns.set(rc={'figure.figsize':(25,5)})
sd = sns.countplot(x='Product_Category', data=dfa)

for bars in sd.containers:
    sd.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
pro_sales=dfa.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='Product_Category', y='Amount', data=pro_sales)
```

```python
pro_ord = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = pro_ord, x = 'Product_ID',y= 'Orders')
```

```python
pro_amn = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = pro_amn, x = 'Product_Category',y= 'Amount')
```
